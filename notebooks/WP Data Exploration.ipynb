```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/wp-police-shootings.csv')
```

```python
data
```

```python
mod = {
    'labels': data['manner_of_death'].unique()
}
mod['left'] = range(len(mod['labels']))
mod['height'] = np.array([ sum(data['manner_of_death'] == x) for x in mod['labels'] ])
```

```python
mod
```

```python
N = len(mod['labels'])
ind = np.arange(N)
width = 0.8
bars = plt.bar(mod['left'], mod['height'])
plt.ylabel('Count')
plt.title('Manner of Death')
plt.xticks(ind + width/2., mod['labels'])
plt.show()
```
